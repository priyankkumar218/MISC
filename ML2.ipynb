# Data Preprocessing on Adult Dataset

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

1. Get the Data

```
col_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'salary']
```

```
dataset = pd.read_csv('https://raw.githubusercontent.com/priyankkumar218/MISC/master/data_pre.csv', names=col_names, na_values=' ?')
```

```
dataset.head(20)
```

```
# numeric_cols = [0, 2, 4, 10, 11, 12]
# categorical_cols = [1, 3, 5, 6, 7, 8, 9, 13, 14]
```

```
X = dataset.iloc[:, 0:14]
y = dataset.iloc[:, -1]
```

```
# Pandas Way
temp = X[['workclass', 'occupation', 'native-country']]
```

```
temp['workclass'].value_counts()
```

```
temp['occupation'].value_counts()
```

```
temp['native-country'].value_counts()
```

```
temp['workclass'] = temp['workclass'].fillna(' Private')
```

```
temp
```

```
# Sklearn Way to fill missing values
from sklearn.impute import SimpleImputer
```

```
sim = SimpleImputer(strategy='most_frequent')
```

```
sim.fit(temp)
```

```
sim.statistics_
```

```
temp = sim.fit_transform(temp)
```

```
temp = pd.DataFrame(temp)
```

```
temp.isnull().sum()
```

```
temp.head(10)
```

```
X[['workclass', 'occupation', 'native-country']] = temp
```

```
X.isnull().sum()
```
